```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Create an example DataFrame using pandas and numpy
# This data could also be loaded from a file, e.g., using pd.read_csv('your_file.csv')
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [20, 34, 30, 35, 25],
    'X_Data': np.random.rand(5) * 100,
    'Y_Data': np.random.rand(5) * 100
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Various Charts using Pandas, Numpy, and Matplotlib')

# --- Bar Chart ---
axes[0].bar(df['Category'], df['Values'], color='skyblue')
axes[0].set_title('Bar Chart: Category Values')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Values')

# --- Pie Chart ---
# The pie chart often works better with sums or proportions of categorical data
sums = df.groupby('Category')['Values'].sum()
axes[1].pie(sums, labels=sums.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Pie Chart: Value Proportions')
# Equal aspect ratio ensures that pie is drawn as a circle
axes[1].axis('equal')

# --- Scatter Plot ---
axes[2].scatter(df['X_Data'], df['Y_Data'], color='coral', marker='o')
axes[2].set_title('Scatter Plot: X vs Y Data')
axes[2].set_xlabel('X_Data')
axes[2].set_ylabel('Y_Data')

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
